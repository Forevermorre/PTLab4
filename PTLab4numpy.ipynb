```python
import numpy as np

print(np.__version__)
```

```python
size = 85

# Заполнение матрицы M
M = 2 * np.ones((size, size), dtype=int)
np.fill_diagonal(M, 2 * np.arange(size) + 3)

# Заполнение матрицы N
N = np.zeros((size, size), dtype=int)

# Четные строки
N[::2, :] = 4
N[::2, 2::3] = 5

# Нечетные строки
N[1::2, :] = 4
N[1::2, 1::2] = 7

p = 1 + np.random.normal(loc=2, scale=2, size=size) # (среднее значение, стандартное отклонение, размер матрицы)

q = np.multiply(N[24, :size], N[71, :size]) # Поэлементное умножение

print('M =\n', M)
print('N =\n', N)
print('p =', p)
print('q =', q)
```

```python
f = np.linalg.det(np.outer(q, q**2).T) # Определитель транспонированной матрицы внешнего произведения q и q^2

g_vector = np.diagonal(M) - np.diagonal(N) + (q**2) * q # Поэлементная сумма разности диагоналей и вектора q^3
g = np.linalg.norm(g_vector, ord=2) # L2 норма

print('f =', f)
print('g =', g)
```
